```python
from plotter import PlotterDataset
from sphere import SphericalDataset
from linear import DatasetGenerator

plotter = PlotterDataset()
```

## Linear Dataset

Generate random points on the surface of an n-dimensional plane

```python
dim_of_space = 2
dim_of_manifold = 1
n_samples = 1000

generator = DatasetGenerator(dim_of_space, dim_of_manifold)
dataset = generator.generate_lines_planes(n_samples, scale=2, noise=0, line_params=[1, 1])
print(f'dataset size: {dataset.shape}')
plotter.plot_sample(dataset)
```

```python
dim_of_space = 3
dim_of_manifold = 2
n_samples = 1000

generator = DatasetGenerator(dim_of_space, dim_of_manifold)
dataset = generator.generate_lines_planes(n_samples, scale=2, noise=0, line_params=[1, 1, 1])
print(f'dataset size: {dataset.shape}')
plotter.plot_sample(dataset)
```

```python
dim_of_space = 3
dim_of_manifold = 1
n_samples = 1000

generator = DatasetGenerator(dim_of_space, dim_of_manifold)
dataset = generator.generate_lines_planes(n_samples, scale=2, noise=0, line_params=[1, 1, 1, 1, 1, 1])
print(f'dataset size: {dataset.shape}')
plotter.plot_sample(dataset)
```

## Spherical Dataset

Generate random points on the surface of an n-dimensional sphere

```python
dim_of_space = 2
dim_of_manifold = 2
n_samples = 1000

generator = SphericalDataset(dim_of_space, dim_of_manifold)
dataset = generator.get_sample(n_samples, radii=[1], noise=0)
print(f'dataset size: {dataset.shape}')
plotter.plot_sample(dataset)
```
